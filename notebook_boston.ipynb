```python
def step_gradient(points, learning_rate, m) :
    
    M = len(points)
    col = points.shape[1]
    m_slope = np.zeros(col)
    for i in range(1, M) :
        x = points[i, 0:-1]
        x = np.append(x, 1)
        y = points[i, -1]
        for j in range(col) :
            m_slope[j] += (-2/M) * (y-(m*x).sum()) * x[j]
    new_m = m - learning_rate * m_slope
    return new_m
```

```python
def cost(points, m) :
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i, 0:-1]
        y = points[i, -1]
        x = np.append(x, 1)
        total_cost += (1/M)*((y - (m*x).sum() )**2)
    return total_cost
```

```python
def gd(points, learning_rate, num_iterations) :
    m = np.zeros(14)
    
    for i in range(num_iterations) :
        m = step_gradient(points, learning_rate, m)
        print(i, "Cost :", cost(points, m))
    return m
```

```python
def run() :
    data = np.genfromtxt("training_boston_x_y_train.csv", dtype = 'float', delimiter = ",")
    learning_rate = 0.08
    num_iterations = 1000
    m = gd(data, learning_rate, num_iterations)
    col = len(data[0])
    return m[0:col-1], m[col-1]
```

```python
def predict(m, x_test, c) :
    y = []
    for i in range(len(x_test)) :
        x = x_test[i,:]
        y.append((m * x).sum() + c)
    return y
#     return m * x_test + c
```

```python
m, c = run()
print(m)
print(c)
```
